##### **Part 1: Orthogonality and the SVD**

**Problem 1.1 (Exam-Style)**

Let $U$ be a square orthogonal real matrix, $\{v_{1},...,v_{n}\}$ its eignvectors and, $\{\lambda_{1},..,\lambda_{n}\}$ its corresponding eigenvalues. Hence, we know that the following holds for any arbitrary eigenvalue and eigenvector:
$$
Uv=\lambda v
$$
If we apply to both sides the Hermitian to both sides we have:
\begin{align*}
    (Uv)^{H}=&(\lambda v)^{H} \\
    v^{H}U^{H}=& v^{H}\lambda^{H} \quad \text{By properties of the Hermitian matrix}\\
    v^{H}\underbrace{U^{H}U}_{I_{n}}v=& v^{H}\lambda^{H}\lambda v \quad \text{Multiplying with the original matrix}\\
    v^{H}I_{n} v=& (\lambda^{H}\lambda)(v^{H}v) \quad \text{Since $U$ is a real matrix}\\
    ||v||^{2}=& (\lambda^{H}\lambda)||v||^{2} \quad \text{$v$ is non-zero}\\
    1=\lambda^{H}\lambda
\end{align*}
Recall that the modulus of a complex number $\lambda=a+bi$ is denoted as $|\lambda|=|a+bi|=\sqrt{a^{2}+b^{2}}$ and $\lambda^{H}\lambda=(a-bi)(a+bi)=a^{2}+b^{2}$ hence, $\lambda^{H}\lambda=|\lambda|^{2}$. Finally, we have:
\begin{align*}
    |\lambda|^{2}=&1\\
    |\lambda|^{2}=&\sqrt{1}
\end{align*}
Since the absolute value of any number is non-negative, we have:
$$
\lambda=1
$$
The eigenvalue is a complex number with absolute value 1. Since we proved for any arbitrary eigenvalue, these holds for all of them. 

**Problem 1.2 (Exam-Style)**

Let's recall that the characteristic polynomial is defined as:
$$
p(\lambda)=det(U-\lambda I)
$$
where $U$ is a square orthogonal $n\times n$ real matrix with $n$ odd. The degree of the polynomial is $n$, so we have $n$ eigenvalues and, let's recall, that the determinant of an orthogonal matrix is: 
$$
det(U)=\pm1=\gamma
$$
We have two cases two consider:

1.- All of the eigenvalues are real numbers. 

2.- Some of the eigenvalues are complex conjugate values and some other are real values. 
Let's recall that:

$$
det(U)=\pm 1=\gamma=\Pi_{i=1}^{n}\lambda_{i} \quad \quad (1)
$$

Using (1), we can make the analysis of our two cases:

  1.- If we have all real eigenvalues, then we have a combination of only 1s or -1s, or a combination of both, to get $\gamma$. So, in either case we have at least a $\pm1$.


  2.- If we have pairs of complex conjugate values ($\lambda_{j}, \lambda_{j}^{H}$), in the limit case we have $\frac{n-1}{2}$ conjugates pairs and, by the proof in 1.1, we know that $\lambda_{j}^{H}\lambda_{j}=|\lambda_{j}|^{2}=1$. Hence, the only left real eigenvalue is either 1 or -1, depending on $\alpha$.


**Problem 1.3 (Exam-Style)**

Let's recall that an square orthogonal matrix has the following property:
$$
R^{T}R=RR^{T}=I
$$
Where $I$ is the identity matrix in a $3\times3$ dimension in our case. If we calculate the determinant for a square orthogonal matrix, we have the following:

\begin{align*}
    det(I)=1=&det(R^{T}R)\\
    =&det(R^{T})det(R)\\
    =&det(R)^{2}
\end{align*}

If we solve the above equation we conclude that:
$$
det(R)=\pm 1
$$
Under our setting the $det(R)=1$, which means that $R$ is a rotation matrix. Now, let's prove that for any matrix we have
$$
Rx=x \quad \quad \text{(2)}
$$
where $x$ is an axis of rotation for $R$. We can re-write (2) as:
$$
(R-I)x=0
$$
We can see that $x$ has to be an eigenvector of the matrix $R$ with the eigenvalue $\lambda=1$.

In problem 1.1 we proved that $|\lambda|=1$. So, we only need to prove that under this setting we have one of the eigenvalues of $R$ is $\lambda=1$. And, this is equivalent to show, y the definition of the characteristic polynomial, that $det(R-I)=0$.

\begin{align*}
    det(R-I)=&det((R-I)^{T})\\
            =&det(R^{-1}-R^{-1}R)\\
            =&det(R^{-1}(I-R))\\
            =&det(R^{-1})det(-(R-I))\\
            =&det(-(R-I))\\
            =&-det(R-I)
\end{align*}

Then, to hold the equality above we need to have that $det(R-I)=0$.

Concluding, we proved that there exists at least one non-zero vector ($x$), such that its eigenvalue is $\lambda=1$ and

$$
(R-I)x=0 \Leftrightarrow Rx=x
$$

$$
U^{T}U=UU^{T}=I
$$

Also, the square root of the eigenvalues of $Q=U^{T}U=I$ are the the singular values of our matrix $U$. And we know that the eigenvalues of the identity matrix are equal to 1.

```
def times(X,Y):
    times = []
    for i in list(range(len(X))):
        a = X[i][0]*Y[0][0]+X[i][1]*Y[1][0]
        b = X[i][0]*Y[0][1]+X[i][1]*Y[1][1]
        times.append([a,b])
    return times
```

```
def transpose(X):
    X_t = [[X[0][0], X[1][0]],[X[0][1],X[1][1]]]
    return X_t
```

```
def eigenvalues(X):
    b = -(X[0][0]+X[1][1])
    c = X[0][0]*X[1][1]-X[0][1]*X[1][0]
    eigen_1 = (-b + sqrt(b*b-4*1*c))/(2*1)
    eigen_2 = (-b - sqrt(b*b-4*1*c))/(2*1)
    eigen_values = [eigen_1,eigen_2]
    return eigen_values
```

```
def eigenvector(X, lambdas):
    small = 1.e-10
    eig_vec = []
    if abs(lambdas[0]-lambdas[1]<small):
        return [[1,0],[0,1]]
    for i in lambdas:
        norm = sqrt((-X[0][1])**2+(X[0][0]-i)**2)
        if norm!=0:
            eig_vec.append([X[0][1]/norm,(-X[0][0]+i)/norm])
        else:
            norm = sqrt((X[1][1]-i)**2+(-X[1][0])**2)
            eig_vec.append([(-X[1][1]+i)/norm,X[1][0]/norm ])
    return transpose(eig_vec)
```

```
def image_to_matrix(image, log_k):
    dim_image = image.shape
    k = 2**(log_k)
    A = np.empty((0,k*k))
    for i in np.arange(0, dim_image[0], k).tolist():
        for j in np.arange(0, dim_image[0], k).tolist():
            row_up = i
            row_down = i + k
            column_left = j
            column_right = j+k
            block = image[row_up:row_down, column_left:column_right]
            vector_block = np.transpose(block.flatten())
            A = np.append(A, [vector_block], axis=0) 
    return A   
```

```
def matrix_to_image(matrix):
    dim_matrix = matrix.shape
    dim_image = [int(sqrt(dim_matrix[0]*dim_matrix[1])),int(sqrt(dim_matrix[0]*dim_matrix[1]))]
    dim_window = [int(sqrt(dim_matrix[1])), int(sqrt(dim_matrix[1]))]
    windows = int(sqrt((dim_matrix[0]*dim_matrix[1])/(dim_window[0]*dim_window[1])))
    image = np.empty((dim_image))
    count = 0
    for i in range(windows):
        for j in range(windows):
            vec = list(matrix[count,:])
            block = []
            for p in range(dim_window[0]):
                lis = vec[p*dim_window[0]:(p+1)*dim_window[0]]
                block.append(lis)
            block = np.array(block, dtype = np.uint8)
            image[0+i*dim_window[0]: dim_window[0]+i*dim_window[0],  0+j*dim_window[1]: dim_window[1]+j*dim_window[1]] = block
            count += 1
    image =  np.array(image,dtype=np.uint8)
    return image 
```

```
from sklearn.decomposition import PCA
from types import SimpleNamespace

def full_image_pca(image, log_k):
    matrix = image_to_matrix(image, log_k)
    pca = PCA(svd_solver='full')
    pca.fit(matrix)
    return SimpleNamespace(mean=pca.mean_, sigmas=pca.singular_values_,
                           U=pca.components_, UT=np.transpose(pca.components_),
                           log_k=log_k)


def transformations(pca, n_components):
    mean = pca.mean
    forward = pca.UT[:, :n_components]
    backward = pca.U[:n_components, :]
    log_k = pca.log_k

    def compressor(image):
        compressed = image_to_matrix(image, log_k)
        return np.dot(compressed - mean, forward)

    def decompressor(matrix):
        decompressed = np.dot(matrix, backward) + mean
        return matrix_to_image(decompressed)

    return compressor, decompressor
```
